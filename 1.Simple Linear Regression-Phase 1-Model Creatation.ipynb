```python
#pandas is python library working with dada sets.....
import pandas as pd
```

```python
#Read the data sets......
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
#saparate the independent variable.......
independent=dataset[["YearsExperience"]]
```

```python
independent
```

```python
#saparate the dependent variable......
dependent=dataset[["Salary"]]
```

```python
dependent
```

```python
#sklearn is python library using data analysis in ML..........
#call train_test_split funtion......
from sklearn.model_selection import train_test_split
#x=input Y=output for given data
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
# imporantand step we use all Algarithom in this same procedure only change algarithom name........

#we using LinearRegression class in sklearn library.......
from sklearn.linear_model import LinearRegression

#we using regression classfication so we assing regressor variable name below.......
regressor=LinearRegression()   #<<<< this is consider class....
regressor.fit(x_train,y_train) #<<<< fit is funtion to find "W" and "B" value........
```

```python
#show the "W"
weight=regressor.coef_ #<<<< coef_ is call funtion
weight
```

```python
bais=regressor.intercept_ #<<<< intercept is call funtion
bais
```

```python
# test the model.......
y_pred=regressor.predict(x_test) #<<<predict is skl function.....
```

```python
from sklearn.metrics import r2_score #<<< r2 is a skl function......
r_score=r2_score(y_test,y_pred) #<<< put actual and predict.....
```

```python
r_score #cross ck prosses if 1<output this good model....
                           # 1>output bad model....... 
```

```python
# save the model..........
import pickle #pickle is save model function......
filename="finalized_model_linear.sav" 
```

```python
pickle.dump(regressor,open(filename,"wb")) #read the file wb is write binary.....
```
